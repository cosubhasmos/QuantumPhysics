## Quantum Harmonic Oscillator

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.special import hermite
```

* $\psi_n(\zeta) = A_n \exp(-\zeta^2/2)H_n(\zeta)$

* $A_n = \sqrt{\dfrac{1}{\sqrt{\pi}2^n n!}}$
* $\zeta = \sqrt{\dfrac{m\omega}{\hbar}}z$

### Deifning initial values

```python
n_values = [0,1, 2, 3]
omega = 1
m = 1
hbar_value = 1
```

### Let's define functions of $\psi$ and potential $V$.

$$\psi_n(z) = A_n \exp(-\dfrac{\zeta^2}{2}) H_n(\zeta)$$
$$V = \dfrac{1}{2}m\omega^2z^2

```python

def psi_n(z, n, m, omega, hbar_value):
    norm_factor = (1/np.sqrt(2**n * math.factorial(n))) * (m*omega / (np.pi*hbar_value))**(1/4)
    hermite_poly = hermite(n)
    return norm_factor * np.exp(-m*omega*z**2 / (2*hbar_value)) * hermite_poly(np.sqrt(m*omega/hbar_value) * z)

def potential_V(z, m, omega):
    return 0.5 * m * omega**2 * z**2
```

Defining interval for $z$

```python
z = np.linspace(-4, 4, 500)
V_z = potential_V(z, m, omega)
```

Finally plotting the wavefunctions for $n = 0, 1, 2, 3$

```python

plt.figure(figsize=(8, 6))
plt.plot(z, V_z, 'r--', label=r'$V(z) = \frac{1}{2} m \omega^2 z^2$')  # Potential

for n in n_values:
    psi = psi_n(z, n, m, omega, hbar_value)
    plt.plot(z, psi + (n+0.5), label=f'n = {n}')

plt.title('Quantum Harmonic Oscillator Eigensolutions')
plt.xlabel('$z$')
plt.ylabel(r'$\psi_n(z)$ and $V(z)$')
plt.legend()
plt.grid(True)
plt.savefig('qho_eigensolutions.png')
plt.show()
```
